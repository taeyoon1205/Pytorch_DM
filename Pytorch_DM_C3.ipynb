## 8절

```python
# 저장 공간을 인데스로 접근하기 위해 사이즈, 오프셋, 스트라이드 정보에 의존
# 오프셋은 저장 공간에서 첫 포인트와의 위치 차이
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
second_point = points[1]
second_point.storage_offset()
```

```python
# 사이즈는 차원 크기
second_point.size()
```

```python
second_point.shape
```

```python
# 스트라이드는 각 차원에서 인덱스를 하나 증가했을 때 저장 공간 상에서 몇 개 요소를 건너뛰어야 하는지에 대한 값
points.stride()
```

```python
# 2차원 텐서에서 요소 i, j에 접근한다면 저장 공간 상으로는 storage_offset + stride[0] * i + stride[1] * j 번째 요소
# 텐서를 통해 스토리지를 간접 접근하는 식으로 텐서의 Transpose(전치)나 일부만으로 더 작은 텐서 만드는 데 필요한 연산을 메모리 재할당을 하지 않으므로 저렴하게 수행 가능
```

```python
# 특정 부분만 정의하여 서브텐서 제작 및 오프셋, 사이즈, 스트라이드 값 확인
second_point = points[1]
second_point.size()
```

```python
second_point.storage_offset()
```

```python
second_point.stride()
```

```python
# points텐서보다 작은 차원을 가지지만 동일한 저장 공간을 가르키므로 second_point텐서를 변경하면 points텐서도 변경됨
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
second_point = points[1]
second_point[0] = 10.0
points
```

```python
# 이런 동작 방식이 항상 옳지는 않으므로 서브텐서를 새 텐서로 복제
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
second_point = points[1].clone()
second_point[0] = 10.0
points
```

```python
# 2차원 텐서들에 대해 transpose를 사용하거나 t 함수 사용하여 전치
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
points
```

```python
points_t = points.t()
points_t
```

```python
# points와 points_t는 같은 공간을 가르킴
print(id(points.storage()) == id(points_t.storage())) # python id함수의 변경으로 인해 False 출력되는 것으로 추정
print(points.storage())
print(points_t.storage())
```

```python
# points와 points_t는 차원 정보와 스트라이드만 다르므로 다음과 같이 출력됨
print(points.stride(), points_t.stride())
```

```python
# 텐서에서 각 요소는 행 단위로 나뉘어짐. Ex)points[0, 0]에서 points[1, 0] 이동은 요소 두 개 이동하는 것이고 points[0, 0]에서 points[0, 1] 이동은 요소 한 개 이동
```

```python
# 다차원 배열에서도 차원 정보와 스트라이드 바뀔 두 차원 지정하면 전치됨
some_t = torch.ones(3, 4, 5)
transpose_t = some_t.transpose(0, 2)
print(some_t.shape, transpose_t.shape)
print(some_t.stride(), transpose_t.stride())
```

```python
# 가장 오른쪽 차원에서 시작해서 증가되는 형태로 저장소에 값이 펼쳐진 텐서는 contiguous로 정의
# 인접한(contiguous) 텐서는 값 순회 시 띄엄띄엄 참조하지 않아 데이터 지역성 관점에서 CPU 메모리 접근 효율이 좋음
# 실제 효율 여부는 사용하는 알고리즘에 따라 다를 수 있음
```

```python
# 파이토치 텐서 연산 중 인접한 텐서에만 동작하는 경우가 있음 ex) view
# 이 경우 경고성 예외를 던져 명시적으로 contiguous를 호출하라고 알려줌
# 텐서가 이미 인접한 상태라면 contiguous가 실제로 하는 일은 없고 성능에도 영향 없음
points.is_contiguous()
```

```python
points_t.is_contiguous()
```

```python
# 텐서는 device 인자로 위치 지정 가능
points_gpu = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]], device = 'cuda')
```

```python
# to 메소드 활용 가능
points_gpu = points.to(device='cuda')
```

```python
# 이런 텐서가 위치하는 곳은 GPU의 RAM
# 보통 텐서 위치에 관계없이 동일한 사용자 API를 제공하므로 연산이 일어나는 장소에 상관없이 코드 작성이 쉬워짐
# GPU 둘 이상이면 0부터 시작하는 정수로 지명
points_gpu = points.to(device='cuda:0') # 이제부터 텐서 연산은 GPU에서 수행
points = 2 * points # CPU에서 실행
points_gpu = 2 * points.to(device='cuda') # GPU에서 실행
```

```python
# 계산 후 points_gpu 텐서는 CPU 영역으로 이동하지 않음
# 위 코드는 아래와 같은 단계 실행
# points 텐서를 GPU에 복사 -> GPU에서 새 텐서 할당 후 곱셈 결과 저장 -> GPU 텐서 핸들을 반환
```

```python
points_gpu = points_gpu + 4 # 연산을 GPU에서 수행하며 출력하거나 결과 텐서 값에 접근하지 않으면 CPU에는 아무 정보도 전달되지 않음
```

```python
# 텐서 CPU로 옮기려면 to 메소드에 cpu 인자 전달
points_cpu = points_gpu.to(device='cpu')
```

```python
# to 메소드 대신 cpu나 cuda라는 축약 메소드 사용 가능
points_gpu = points.cuda() # default = 0
print(points_gpu.device)
points_gpu = points.cuda(0)
print(points_gpu.device)
points_cpu = points_gpu.cpu()
print(points_cpu.device)
```
